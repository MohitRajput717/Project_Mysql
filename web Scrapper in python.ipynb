```python
from autoscraper import AutoScraper
```

```python
f_url="https://www.flipkart.com/search?q=ipnone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
wanted_list=["₹48,000","APPLE iPhone 11 (Black, 128 GB)","4.680,989 Ratings & 6,047 Reviews"]
```

```python
scraper=AutoScraper()
result=scraper.build(f_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(f_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_9386':'Title','rule_p3fs':'price','rule_met3':'reviews'})
scraper.keep_rules(['rule_9386','rule_p3fs','rule_met3'])
scraper.save('fk-search')
```

```python
results=scraper.get_result_similar('https://www.flipkart.com/search?q=phone%20under%2012k&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off',group_by_alias=True)
```
