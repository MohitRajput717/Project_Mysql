Q-13. Write an SQL query to show only odd rows from a table.

Ans. select *
from
(
SELECT *, ROW_NUMBER() OVER(ORDER BY (select 1)) AS rownumber
FROM contacts
)
contacts
where contacts.RowNumber % 2 = 1;
