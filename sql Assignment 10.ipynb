 Write an SQL query to determine the 5th highest salary without using TOP or limit method.

Ans. SELECT first_name, last_name, deptartment, salary
FROM (
SELECT first_name, last_name, deptartment, salary, ROW_NUMBER() over( order by salary desc) as highest_salary
FROM worker_fil
	)
worker_fil
WHERE highest_salary <= 5;

Q-20. Write an SQL query to fetch the list of employees with the same salary.

Ans.
select *
from worker_fil
where salary in (select salary from worker_fil group by salary having count(1)>1)
order by salary;
